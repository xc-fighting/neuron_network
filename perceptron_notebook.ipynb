```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lines = None
with open('./data.csv') as f:
    lines = f.readlines()
x = []
y = []
flag = []
for line in lines:
    arr = line.split(',')
    x.append(float(arr[0]))
    y.append(float(arr[1]))
    flag.append(float(arr[2]))
#print(x)
    
```

```python
size = len(x)
for idx in range(size):
    pos_x = x[idx]
    pos_y = y[idx]
    flag_val = flag[idx]
    #print("{} {} {}".format(pos_x, pos_y, flag_val))
    if flag_val == 1.0:
        plt.scatter(pos_x, pos_y, color='green')
    else:
        plt.scatter(pos_x, pos_y, color='blue')

historys = None
with open('./model.txt') as fp:
    historys = fp.readlines()
    
candidate_error = 65535
candidate_w1 = -1
candidate_w2 = -1
candidate_bias = -1

for history in historys:
    arr = history.split(',')
    w1 = float(arr[0])
    w2 = float(arr[1])
    bias = float(arr[2])
    error = int(arr[3])
    if error < candidate_error:
        candidate_error = error
        candidate_w1 = w1
        candidate_w2 = w2
        candidate_bias = bias
        
print("{} {} {}".format(candidate_w1, candidate_w2, candidate_bias))
print("error:{}".format(candidate_error))       
line_x = [0, -(candidate_bias/candidate_w2)]
print("point1:" + str(line_x))
line_y = [-(candidate_bias/candidate_w1), 0]
print("point2:" + str(line_y))
plt.plot(line_x, line_y)
plt.show()
```
